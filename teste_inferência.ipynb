```python
import statsmodels.api as sm
# Definindo a variável independente (VI) e dependente (VD)
X = df['HEI2015_TOTAL_SCORE']
y = df['depressao']

# Adicionando uma constante ao modelo (intercepto)
X = sm.add_constant(X)

# Ajustando o modelo de regressão linear
model = sm.OLS(y, X).fit()

# Exibindo o resumo dos resultados
print(model.summary())
```

```python
import scipy.stats as stats
filtered_df = df[df['DPQ010'].isin([0, 1, 2, 3])]

# Separando os dados em grupos com base nas categorias de DPQ010
group0 = filtered_df[filtered_df['DPQ010'] == 0]['HEI2015_TOTAL_SCORE']
group1 = filtered_df[filtered_df['DPQ010'] == 1]['HEI2015_TOTAL_SCORE']
group2 = filtered_df[filtered_df['DPQ010'] == 2]['HEI2015_TOTAL_SCORE']
group3 = filtered_df[filtered_df['DPQ010'] == 3]['HEI2015_TOTAL_SCORE']

# Realizando o teste ANOVA
f_statistic, p_value = stats.f_oneway(group0, group1, group2, group3)

# Exibindo os resultados
print(f'Estatística F: {f_statistic}')
print(f'Valor p: {p_value}')
```

```python
# Filtrando as categorias desejadas para DPQ010 e removendo valores NaN
filtered_df = df[df['DPQ010'].isin([0, 1, 2, 3])]
filtered_df = filtered_df.dropna(subset=['HEI2015_TOTAL_SCORE', 'DPQ010'])

# Separando os dados em grupos com base nas categorias de DPQ010
group0 = filtered_df[filtered_df['DPQ010'] == 0]['HEI2015_TOTAL_SCORE']
group1 = filtered_df[filtered_df['DPQ010'] == 1]['HEI2015_TOTAL_SCORE']
group2 = filtered_df[filtered_df['DPQ010'] == 2]['HEI2015_TOTAL_SCORE']
group3 = filtered_df[filtered_df['DPQ010'] == 3]['HEI2015_TOTAL_SCORE']

# Verificando se algum grupo está vazio
if group0.empty or group1.empty or group2.empty or group3.empty:
    print("Um ou mais grupos estão vazios. Não é possível realizar o teste ANOVA.")
else:
    # Realizando o teste ANOVA
    f_statistic, p_value = stats.f_oneway(group0, group1, group2, group3)

    # Exibindo os resultados
    print(f'Estatística F: {f_statistic}')
    print(f'Valor p: {p_value}')
```
