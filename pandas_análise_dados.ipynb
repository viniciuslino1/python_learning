Loc: Labes

Loc: procurando valores no index ou através de operações e máscaras com boleanos. Ao contrário do iloc que busca selecionar linhas e colunas por seu index numérico, o operador .loc, busca por valores que podem ser strings ou números dentro de colunas e indexes.

```python
#com operações boleanas
# Usado para buscar valores específicos em uma coluna e selecionar o dataset a partir dela. Essa não é a única forma de fazer seleções a partir de um único valor.
df.loc[df['Sex'] == 'female'].head()
```

LAMBDA

```python
df_name = df.copy()
```

```python
#criando uma nova coluna chamada last name com base na coluna name usando lambd
df_name['Last_name'] = df_name['Name'].apply(lambda x: x.split(",")[0].replace('', ""))
```

```python
df_name.head()
```

```python
#quando tem uma palavra escrita errada em uma coluna, podemos passar um dicionário para substituir ela na coluna
df_name['Last_name'].replace({'Heikkinen':'Heineken', 'Allen':'Alien'}, inplace=True)
```

```python
df_name.head()
```

```python
#criando uma função e aplicando com apply: substituir as letras dos portos de embarque por seus nomes
def port_emkd(x):
    if x == 'S':
        x = 'Southampton'
    if x == 'C':
        x = 'Cherbourg'
    if x == 'Q':
        x = 'Queenstown'
    return x
```

```python
df_name['Embarked'] = df_name['Embarked'].apply(lambda x: port_emkd(x))
```

```python
df_name.head()
```

```python
#transformando categóricas, o pandas vai retornar códigos para cada categoria

df_name['c_Embarked'] = df_name['Embarked'].astype('category').cat.codes
df_name.head(13)
```

```python
#gerar gráfico de musicas por ano
import matplotlib.pyplot as plt
```

```python
df_novo_burn['data'] = pd.to_datetime(df_novo_burn['data'], format='mixed')
```

```python
df_novo_burn['ano'] = df_novo_burn['data'].dt.year
```

```python
df_novo_burn.head()
```

```python
contagem_musicas_por_ano = df_novo_burn['ano'].value_counts().sort_index()
```

```python
plt.figure(figsize=(10, 6))
contagem_musicas_por_ano.plot(kind='bar')
plt.xlabel('Ano')
plt.ylabel('Número de Músicas')
plt.title('Contagem de Músicas por Ano')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
ax = contagem_musicas_por_ano.plot(kind='bar')
plt.xlabel('Ano')
plt.ylabel('Número de Músicas')
plt.title('Contagem de Músicas por Ano')
plt.xticks(rotation=0)

# Adicione rótulos dos dados em cima das barras
for container in ax.containers:
    ax.bar_label(container, label_type='edge')

plt.show()
```
