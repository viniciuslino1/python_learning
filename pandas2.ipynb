#Visualização de histogramas com Pandas

A biblioteca Pandas também nos permite utilizar recursos de visualização de dados. Uma das visualizações que pode ser bastante útil em um processo de análise exploratória é o histograma

```python
import pandas as pd
#criando data set
data = {
    'idades':[10,25,24,42,50,12,22,37,88],
    'altura':[165,180,176,184,160,155,170,167,158]

}
#transformando em data frame
df = pd.DataFrame(data)
_ = df.hist(bins=5)
```

```python
#plotar somente hitograma das idades
_ = df['idades'].hist(bins=5)
```
